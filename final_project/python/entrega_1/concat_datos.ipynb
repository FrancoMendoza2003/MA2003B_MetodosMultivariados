## NORTE2, NOROESTE2, SURESTE3

```python
df_norte2 = pd.concat(
    [
        pd.read_excel(
            "../00_raw_datasets/DATOS_HISTORICOS_2020_2021_TODAS_ESTACIONES.xlsx",
            sheet_name="NORTE2",
            parse_dates=["date"],
        ),
        pd.read_excel(
            "../00_raw_datasets/DATOS_HISTORICOS_2022_2023_TODAS_ESTACIONES.xlsx",
            sheet_name="NORTE2",
            parse_dates=["date"],
        ),
    ]
).sort_values("date")

df_norte2.head()
```

```python
df_noroeste2 = pd.concat(
    [
        pd.read_excel(
            "../00_raw_datasets/DATOS_HISTORICOS_2020_2021_TODAS_ESTACIONES.xlsx",
            sheet_name="NOROESTE2",
            parse_dates=["date"],
        ),
        pd.read_excel(
            "../00_raw_datasets/DATOS_HISTORICOS_2022_2023_TODAS_ESTACIONES.xlsx",
            sheet_name="NOROESTE2",
            parse_dates=["date"],
        ),
    ]
).sort_values("date")

df_noroeste2.head()
```

```python
df_sureste3 = pd.concat(
    [
        pd.read_excel(
            "../../datos/00_raw_datasets/DATOS_HISTORICOS_2020_2021_TODAS_ESTACIONES.xlsx",
            sheet_name="SURESTE3",
            parse_dates=["date"],
        ),
        pd.read_excel(
            "../../datos/00_raw_datasets/DATOS_HISTORICOS_2022_2023_TODAS_ESTACIONES.xlsx",
            sheet_name="SURESTE3",
            parse_dates=["date"],
        ),
    ]
).sort_values("date")

df_sureste3.head()
```
